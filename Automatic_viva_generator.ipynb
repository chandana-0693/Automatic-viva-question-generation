```python
from transformers import pipeline
import nltk
nltk.download('punkt')

# Use the CORRECT pipeline for T5
qg = pipeline("text2text-generation", model="valhalla/t5-small-qg-prepend")

text = """
Natural Language Processing is a field of Artificial Intelligence.
It helps computers understand, interpret and generate human language.
NLP is used in chatbots, translation systems and voice assistants.
"""

input_text = "generate questions: " + text

output = qg(input_text, max_length=128)

print("\nGenerated Questions:\n")
print(output[0]['generated_text'])
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("valhalla/t5-small-qg-prepend")
model = AutoModelForSeq2SeqLM.from_pretrained("valhalla/t5-small-qg-prepend")

# Your input text
text = """
Machine Learning is a subset of Artificial Intelligence (AI) where computers learn patterns from data and make predictions or decisions without being explicitly programmed.
"""

input_text = "generate questions: " + text

# Tokenize input
inputs = tokenizer.encode(input_text, return_tensors="pt")

# --- REPLACE THIS LINE ---
# outputs = model.generate(inputs, max_length=512)
# WITH MULTI-QUESTION GENERATION:
outputs = model.generate(
    inputs,
    max_length=512,       # increase length so model can generate more
    num_return_sequences=3,  # generate 3 different sequences
    do_sample=True,       # sample for variety
    top_k=50,
    top_p=0.95
)

# Decode and print each sequence
print("\nGenerated Questions:\n")
for i, out in enumerate(outputs):
    questions = tokenizer.decode(out, skip_special_tokens=True)
    print(f"Question Set {i+1}:\n{questions}\n")
```
